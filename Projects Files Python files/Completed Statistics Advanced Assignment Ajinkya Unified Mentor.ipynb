# Task 4 - Obtain the logical conditions for the features "Scr_h" and "Scr_w"

```python
# Let's tackle these features: "Scr_h", "Scr_w"import matplotlib.pyplot as plt
import seaborn as sns

# Plot Scr_h
sns.histplot(df["Scr_h"], kde=True)
plt.title("Distribution of Screen Height (Scr_h)")
plt.xlabel("Scr_h")
plt.ylabel("Frequency")
plt.show()

# Plot Scr_w
sns.histplot(df["Scr_w"], kde=True)
plt.title("Distribution of Screen Width (Scr_w)")
plt.xlabel("Scr_w")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create a new feature called "Scr_d" which stores the length of the diagonal of the screen of the phone
# Create a new feature for screen diagonal using Pythagorean theorem
df["Scr_d"] = np.sqrt(df["Scr_h"]**2 + df["Scr_w"]**2)

# Preview the new feature
print(df[["Scr_h", "Scr_w", "Scr_d"]].head())

```

```python
# Create a histogram of the "Scr_d" feature and also show the quartiles
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate quartiles
q1 = df["Scr_d"].quantile(0.25)
q2 = df["Scr_d"].median()
q3 = df["Scr_d"].quantile(0.75)

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df["Scr_d"], kde=True, bins=15)

# Overlay the quartiles
plt.axvline(q1, color='purple', linestyle='--', label=f'Q1: {q1:.2f}')
plt.axvline(q2, color='green', linestyle='-', label=f'Median (Q2): {q2:.2f}')
plt.axvline(q3, color='orange', linestyle='--', label=f'Q3: {q3:.2f}')

# Formatting the plot
plt.title("Distribution of Screen Diagonal (Scr_d)")
plt.xlabel("Screen Diagonal Length")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# The children want phones that have very good screen sizes
# Consider the phones that have screen sizes greater than or equal to the upper quartile value in the data set
# Create a logical condition for this situation and store the logical values as "con3"

# Calculate the upper quartile (Q3) of screen diagonal
q3_scr_d = df["Scr_d"].quantile(0.75)

# Create the logical condition
con3 = df["Scr_d"] >= q3_scr_d

# Preview the logical values
print(con3.head())


```

Observations:

The features "Scr_h" and "Scr_w" are respectively the height and the width of the phone screen.

We created a new feature called "Scr_d" which is essentially the length of the screen diagonal.

The upper quartile has been selected as a threshold in this case as the children were very particular on this point.

In case it is too strict, we can choose the mean or the median as a threshold.

# Task 9 - Study the variability of the features in the original data set

```python
# Calculate the ratio of the standard deviation to the mean for all the numerical features in the dataframe
# Store these values in a new series wherein the rows are the features and the only column is the calculated ratio
# Name the series as "deviations"

# Select only numerical columns
numerical_df = df.select_dtypes(include=[np.number])

# Calculate the standard deviation to mean ratio
deviations = (numerical_df.std() / numerical_df.mean()).to_frame(name='deviations')

# Display the result
print(deviations)
```

```python
# View the "deviations" series after sorting it in descending order
# Sort the deviations series in descending order
deviations_sorted = deviations.sort_values(by="deviations", ascending=False)

# Display the sorted result
print(deviations_sorted)

```

Observations:
The ratio of the standard deviation to the mean of a feature normalizes its scale, allowing for meaningful comparison across different units and magnitudes.

This enables us to assess which features are relatively more or less variable.

The most variable feature in the original data set is Int_Mem.

The least variable feature in the original data set is Px_w.

The ratio of standard deviation to mean for the Price feature is 0.74, indicating that phone prices are indeed quite variable.

This variability in price, despite lower variability in most other features, suggests that other unmeasured or compounded factors may be influencing pricing.
